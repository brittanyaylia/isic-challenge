```python
import numpy as np
import pandas as pd
import os
from shutil import copyfile
```

## Add more data (from 2020) to train into the original 2016 ISIC dataset.
### Do not run cells. The second dataset of images and labels (https://www.kaggle.com/c/siim-isic-melanoma-classification) are used to add to the original dataset, and they are not committed to the repo. The combined images and labels are committed.

```python
# This file will not be in repo
df = pd.read_csv('./Labels/train.csv') 
```

```python
# Filter for all malignant labels.
df_new_mag = df[df['benign_malignant']=='malignant']

# Add the ID names of the images that are labeled "malignant" to a list
mag_files = df_new_mag['image_name'].tolist()
src_dir ='./Images/train_images' # this file will not be in repo
dst_dir = './Images/ISBI2016_ISIC_Part3_Training_Data'

# copy all the malignant images to the old training folder
for f in mag_files:
    copyfile(f'{src_dir}/{f}.jpg', f'{dst_dir}/{f}.jpg')
```

```python
# Sort the names so they are in order
df_new_mag.sort_values('image_name')
```

```python
# Filter for all benign labels. Dataset is large, will only take the first 700.
df_new_ben = df[df['benign_malignant']=='benign'][:700]

# Add the ID names of the images that are labeled "benign" to a list
ben_files = df_new_ben['image_name'].tolist()
src_dir ='./Images/train_images' # this file will not be in repo
dst_dir = '../Images/ISBI2016_ISIC_Part3_Training_Data'

# copy all the benign images to the old training folder
for f in ben_files:
    copyfile(f'{src_dir}/{f}.jpg', f'{dst_dir}/{f}.jpg')
```

```python
# Sort the names so they are in order
df_new_ben.sort_values('image_name')
```

```python
# Reorder the labels
df_n_mag2 = df_new_mag[['image_name', 'benign_malignant']].set_index('image_name')
df_n_ben2 = df_new_ben[['image_name', 'benign_malignant']].set_index('image_name')

# Set index of the original labels to match the index of the new labels
df_old = pd.read_csv('./Labels/ISBI2016_ISIC_Part3_Training_GroundTruth.csv', header=None) # this file will not be in repo
df_old.columns = ['image_name', 'benign_malignant']
df_old_2 = df_old.set_index('image_name')

# Combine the new labels with the old labels
df_combo = pd.concat([df_old_2, df_n_mag2, df_n_ben2])
df_combo.sort_index(inplace=True)
df_combo.to_csv('./Labels/train_combined_GT.csv')
```
