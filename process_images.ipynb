```python
import os 
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Process Training Images (900)

```python
# sample one image, 0 for grayscale
image = cv2.imread('./Images/ISBI2016_ISIC_Part3_Training_Data/ISIC_0000000.jpg', 0)
plt.imshow(image)
```

```python
image.shape
```

```python
# for each image in Training Data folder, read the image, append the size of the image to list imgshapes
imgshapes = []
for imgpath in os.scandir('./Images/ISBI2016_ISIC_Part3_Training_Data/'):
    img = cv2.imread(imgpath.path, 0)
    imgshapes.append(img.shape)
```

```python
min(imgshapes)
```

```python
max(imgshapes)
```

```python
# set desired size. For each image in folder, read the image, resize to desired size, save image to folder Resized_Training_Data
dsize = (480, 360)
for imgsize in os.scandir('./Images/ISBI2016_ISIC_Part3_Training_Data/'):
    img = cv2.imread(imgsize.path, 0)
    imgresized = cv2.resize(img, dsize)
    cv2.imwrite(f'./Images/Resized_Training_Data/{imgsize.name}', imgresized)
```

```python
# view sample of a resized image
plt.imshow(imgresized)
```

```python
# for each resized image in folder, read image and append it to list xtrain
xtrain = []
for imgresized in os.scandir('./Images/Resized_Training_Data/'):
    image = cv2.imread(imgresized.path, 0)
    xtrain.append(image)
```

```python
# convert list to np array
xtrain_nparray = np.array(xtrain)
```

```python
# save np array
np.save('./Data/xtraindata.npy', xtrain_nparray)
```
