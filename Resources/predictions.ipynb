```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import cv2
import os
```

```python
# Load model
cnn_model = load_model("cnn_model.h5")
```

### Make Prediction

```python
# Set desired shape. For cv2, it is (height,width)
dsize = (480, 360)

# Append image paths to list
prediction_paths = [img.path for img in os.scandir('./Images/Predictions/')]

# For each path, show the image, read the image in grayscale, resize the image, normalize the pixel values, print prediction

for path in prediction_paths:
    image = cv2.imread(path, 1)
    fig,ax = plt.subplots(1)
    ax.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))   
    image = cv2.imread(path, 0)
    imgresized = cv2.resize(image, dsize)/255
    prediction = cnn_model.predict(imgresized.reshape(1,360,480,1))
    ax.text(image.shape[-1]/2, -20, f'{os.path.basename(path).split(".")[0]}\n{prediction}', va='bottom', ha='center')
#     print(f'Prediction: {prediction}')
```
