```python
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt
import cv2
```

```python
# Load model
cnn_model = load_model("cnn_model.h5")
```

```python
# Make prediction
def prediction(img_path):
    image = cv2.imread(img_path, 0)/255
    dsize = (480, 360)
    imgresized = cv2.resize(image, dsize)
    fig,ax = plt.subplots(1)
    ax.imshow(image)
    return cnn_model.predict(imgresized.reshape(1,360,480,1))
```

```python
prediction('./Images/Predictions/ISIC_0362853-b.jpg')
```
